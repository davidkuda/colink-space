```python
import os

import pandas as pd
```

```python
names_raw = pd.read_csv('VornamenSchweiz.csv', sep=';')
```

```python
names_raw
```

```python
names = names_raw.replace('*', 0).astype({'male': 'int', 'female': 'int'})
```

```python
num = 2000
names[
    (names["male"] > num) |
    (names["female"] > num)
]
```

```python
male_names = (
    names
        .drop("female", 1)
        .loc[names["male"] >= 1900]
        .reset_index(drop=True)
)
```

```python
female_names = (
    names
        .drop("male", 1)
        .loc[names["female"] >= 1900]
        .reset_index(drop=True)
)
```

```python
# https://stackoverflow.com/questions/29517072/add-column-to-dataframe-with-constant-value
normalized_male_names = male_names.rename(columns={"male": "count"}).assign(gender='male')
normalized_female_names = female_names.rename(columns={"female": "count"}).assign(gender='female')
```

```python
top_names = pd.concat([
    normalized_male_names,
    normalized_female_names
]).sort_values('count', ascending=False).reset_index(drop=True)
```

```python
top_names
```

```python
top_names.to_csv('top_names_switzerland.csv', index=False)
```
